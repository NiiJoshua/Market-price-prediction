## Data Visualization

* We are setting the label to be price

### Visualizing categorica variables against price

```python
# function to plot all categorical variables against the label
def plot_box(df, cols, col_y = 'price'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot(col, col_y, data=df)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.xticks(rotation='45')
        plt.show()
```

```python
cat_cols = ['region','market_city','commodity','month','year']
plot_box(df, cat_cols)
```

#### Discusion of results of the box plot

* Prices in the Central region seemed to be higher, followed by Western and Greater Accra
* Drilling down to individual cities, Cape Coast leads followed by Accra and Kumasi
* For comodity, imported rice is higher than the other comodities and plantain has the lowest of prices
* General prices were relatively lowest in October with June recording the highest
* General trends in Ghana sees prices increasing always and this is evident in the last graph

```python
# function to plot all categorical variables against the label
def count_plot(df, cols):
    for col in cols:
        sns.set_style("whitegrid")
        sns.countplot(col, data=df)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel('count')# Set text for y axis
        plt.xticks(rotation='45')
        plt.show()
```

```python
count_plot(df, cat_cols)
```

## Applying time series on the data

```python
# seledting only
ts = df[['year','price']]
ts.head()
```

```python
ts.rename(columns={'year':'y','price':'ds'}, inplace=True)
ts.head()
```

```python
from prophet import Prophet
```

```python
# creating model to train and make predictions
m = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=True)
```

```python
#fitting model
%time m.fit(ts)
```

```python
from prophet.diagnostics import cross_validation
```

```python
cv_results = cross_validation(m,  initial="60 days", horizon="7 days", period="15 days")
```
